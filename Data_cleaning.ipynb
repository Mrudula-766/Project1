```python
df = pd.read_csv('/content/university_towns.txt', sep='\t')
```

```python
university_towns = []
with open('/content/university_towns.txt') as file:
  for line in file:
    if '[edit]' in line:
      #Remember this state until the next found
      state = line
    else:
      #Otherwise, we have a city; keep 'state' as last-seen
      university_towns.append((state, line))
```

```python
university_towns[:5]
```

```python
df
```

```python
towns_df = pd.DataFrame(university_towns,
                        columns=['State', 'RegionName'])
```

```python
towns_df.head()
```

```python
def get_citystate(item):
  if '(' in item:
    return item[:item.find('(')]
  elif '[' in item:
    return item[:item.find('[')]
  else:
    return item
```

```python
towns_df = towns_df.applymap(get_citystate)
towns_df.head()
```
